# Aula 06

## Manipulação de arquivos a partir de uma UI feita com PySide6

A manipulação desses arquivos é feita de forma bastante parecido com o que fizemos na aula passada. A diferença principal é o uso de *widgets* diferentes para visualização e edição. Ao mesmo tempo, nada impede de serem utilizados classes e métodos próprios como [QFile](https://doc.qt.io/qtforpython-6/PySide6/QtCore/QFile.html) e [QJsonDocument](https://doc.qt.io/qtforpython-6/PySide6/QtCore/QJsonDocument.html), dentre outros.
